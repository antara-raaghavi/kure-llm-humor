### Filtering movies.csv

```python
import pandas as pd

def bracket_article_cleanup(dfr):
    #parenthesis removal + "x, The/A/An" -> The X long string comprehension
    fn = lambda x: ' '.join([part.strip() for part in x.split(' (')[0].split(',')][::-1]) if ", The" in x or ", A" in x or ", An" in x else x.split(' (')[0]
    #AH forgot to reassign!
    
    dfr['title'] = dfr['title'].apply(fn)
    return dfr

def filter(csv_file, genre, output):
    df = pd.read_csv(csv_file, encoding='latin1', delimiter='\t')
    
    df['genres'] = df['genres'].apply(lambda x: x.split('|'))
    filtered_df = df[df['genres'].apply(lambda x: genre in x)]


    filtered_df = bracket_article_cleanup(filtered_df)

    with open(output, 'w') as f:
        for title in filtered_df['title']:
            f.write("%s\n" % title)


c = 'movies.csv'
genre = 'Romance'  
output = 'romance_movies.txt'

filter(c, genre, output)
```
